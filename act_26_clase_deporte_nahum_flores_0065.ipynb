```python
class Deporte:
    def __init__(self, nombre, tipo, jugadores):
        # Inicializamos los atributos
        self.nombre = ""
        self.tipo = ""
        self.jugadores = 0

    def leerdatos(self):
        print("--- Ingreso de datos ---")
        self.nombre = input("Ingrese el nombre del deporte: ")
        self.tipo = input("Ingrese el tipo (Individual/Equipo): ")
        self.jugadores = int(input("Ingrese la cantidad de jugadores: "))

    def mostrardatos(self):
        print("\n--- Información del Deporte ---")
        print(f"Nombre: {self.nombre}")
        print(f"Tipo: {self.tipo}")
        print(f"Jugadores: {self.jugadores}")

## fin de la clase Deporte

# Creando los objetos deporte1 y deporte2
print("REGISTRO DE DEPORTES")
print("********************")

# Objeto 1
print("\nDeporte Numero 1")
deporte1 = Deporte("", "", 0)
deporte1.leerdatos()
deporte1.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*")

# Objeto 2
print("\nDeporte Numero 2")
deporte2 = Deporte("", "", 0)
deporte2.leerdatos()
deporte2.mostrardatos()

print("\n-*-*-*-*-*-*-*-*-*-*-*-*")
print("Programa realizado por Nahum Flores")
```
